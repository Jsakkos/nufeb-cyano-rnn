Region I

2500 simulations were done with NUFEB model for sucrose-export cyanobacteria. The simulations were done by varying the inputs, i.e. number of bacteria (n_cyano), amount of CO2 (CO2), amount of light (Light) and sucrose-export rate (SucroseExport), according to the Morris sampling. 

The range of variation for each input was [1,50], [15,100], [0.001,1],[0,1] for n_cyano, CO2, Light) and SucroseExport,  respectively. All the inputs parameters used for the simulations are stored in the file: 50up-input_params.txt

For each simulation, population curves were obtained. Each population curve has a maximum (A), slope (mu) and lag time(tau). These are the ouptuts. All the outputs obtained are stored in the file: 50up-bare-fitting-results.txt

This notebook shows:
1. Histograms for n_cyano, CO2, Light and SucroseExport.
2. Histograms for A, mu and tau.
3. The plots: mu vs. tau, A vs. tau, mu vs. tau and mu vs. tau
4. The plots: A/mu/tau vs. n_cyano, CO2, Light, SucroseExport


```python
# Import required packages
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
%matplotlib inline
import gdown
from scipy.stats import gaussian_kde
```

```python
#Importing the values for the input parameters
!gdown "https://drive.google.com/uc?id=1lqeWShKTksWUI1qhTMHP1yXWx4ugQaYT"
inputs = np.loadtxt('50down-input_params.txt',float)
```

```python
#Plotting the range of input parameters used

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(1,4,1)
ax.set_title('n_cyano')
ax.hist(inputs[:,0], bins = 50);
ax.set_xlabel('n_cyano')
# ax.set_xlim([500, 1500])
# ax.set_ylim([0, 10])
# plt.xticks(fontsize=20)
# plt.yticks(fontsize=20)

ax = fig.add_subplot(1,4,2)
ax.set_title('CO2')
ax.hist(inputs[:,1], bins = 50);
ax.set_xlabel('CO2')
#ax.set_xlim([-4, 4])
#ax.set_ylim([0, .8])

ax = fig.add_subplot(1,4,3)
ax.set_title('Light')
ax.hist(inputs[:,2], bins = 50);
ax.set_xlabel('Light')
#ax.set_xlim([-4, 4])
#ax.set_ylim([0, .8])

ax = fig.add_subplot(1,4,4)
ax.set_title('SucroExport')
ax.hist(inputs[:,3], bins = 50);
ax.set_xlabel('SucrosExport')
# ax.set_ylabel('Count',fontsize=20)
#ax.set_xlim([-4, 4])
#ax.set_ylim([0, .8])
```

```python
#Importing the A, mu, tau, and A0 results obtained from the fitting
!gdown "https://drive.google.com/uc?id=1YDpI7avJCVncEWEEmF_aA7G5kat8GFmh"
outputs = np.loadtxt('50down-bare-fitting-results.txt',float)
```

```python
#Plotting the A, mu and tau values obtained

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(1,3,1)
ax.set_title('A', fontsize=25)
ax.hist(outputs[:,0], bins = 50);
ax.set_xlabel('A',fontsize=20)
ax.set_ylabel('Count',fontsize=20)
#ax.set_xlim([-4, 4])
#ax.set_ylim([0, .8])
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

ax = fig.add_subplot(1,3,2)
ax.set_title('mu', fontsize=25)
ax.hist(outputs[:,1], bins = 50);
ax.set_xlabel('mu',fontsize=20)
ax.set_ylabel('Count',fontsize=20)
#ax.set_xlim([-4, 4])
#ax.set_ylim([0, .8])
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

ax = fig.add_subplot(1,3,3)
ax.set_title('tau', fontsize=25)
ax.hist(outputs[:,2], bins = 50);
ax.set_xlabel('tau',fontsize=20)
ax.set_ylabel('Count',fontsize=20)
#ax.set_xlim([-4, 4])
#ax.set_ylim([0, .8])
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

```python
#Looking for correlations among these output parameters

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(1,3,1)
ax.set_title('mu vs A',fontsize=25) 
ax.plot(outputs[:,0],outputs[:,1],marker='o', linestyle = '')
ax.set_xlabel('A')
ax.set_ylabel('mu')

ax = fig.add_subplot(1,3,2)
ax.set_title('A vs tau',fontsize=25) 
ax.plot(outputs[:,0],outputs[:,2],marker='o', linestyle = '')
ax.set_xlabel('A')
ax.set_ylabel('tau')

ax = fig.add_subplot(1,3,3)
ax.set_title('mu vs tau',fontsize=25) 
ax.plot(outputs[:,1],outputs[:,2],marker='o', linestyle = '')
ax.set_xlabel('mu')
ax.set_ylabel('tau')
```
